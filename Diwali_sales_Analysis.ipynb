```python
# Sample DataFrame
C = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Define a color for each bar, ensuring you have enough colors
colors = sns.color_palette("husl", C['Age Group'].nunique())  # "husl" generates as many colors as there are unique age groups

# Create the bar plot, with hue set to Age Group
sns.set(rc={'figure.figsize':(10, 6)})  # Set the figure size if needed
ax = sns.barplot(x='Age Group', y='Amount', data=C, palette=colors, hue='Age Group', dodge=False, legend=False)

# Add labels to each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9), 
                textcoords='offset points')
```

From the above graph we can consider that most of the buyers are of Age group 26-35 yrs

```python
AB = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc ={'figure.figsize':(25,5)})
sns.barplot(x = 'State',y = 'Orders', data = AB, palette = 'deep')
```

From the above graph we can observe that most of the orders are from the Uttar Pradesh state.

```python
AC = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State', y='Amount', data=AC, palette='deep')
plt.show()

```

From the above graphs most of the sales are from Uttar Pradesh, Maharastra  and Karnataka.
